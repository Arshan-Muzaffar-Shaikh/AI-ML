```python
from collections import deque
import matplotlib.pyplot as plt
import networkx as nx

# BFS function
def bfs(graph, start):
    visited = set()
    queue = deque([start])
    visited.add(start)
    traversal = []

    while queue:
        node = queue.popleft()
        traversal.append(node)
        for child in graph[node]:
            if child not in visited:
                visited.add(child)
                queue.append(child)
    return traversal

# Tree-shaped adjacency list
graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F', 'G'],
    'D': [],
    'E': [],
    'F': [],
    'G': []
}

# Run BFS
order = bfs(graph, 'A')
print("BFS Traversal:", order)

# Create graph in NetworkX
G = nx.DiGraph()
for parent, children in graph.items():
    for child in children:
        G.add_edge(parent, child)

# Define fixed positions for exact tree shape
pos = {
    'A': (0, 3),
    'B': (-1, 2),
    'C': (1, 2),
    'D': (-1.5, 1),
    'E': (-0.5, 1),
    'F': (.5, 1),
    'G': (1.5, 1)
}

# Draw graph
plt.figure(figsize=(6, 4))
nx.draw(G, pos, with_labels=True, node_size=2000, node_color="lightgreen", arrows=True, font_size=12)
plt.title("Tree Structure (BFS)")
plt.show()
```

```python
from collections import deque
import matplotlib.pyplot as plt
import networkx as nx

# DFS function
def dfs(graph, start, visited=None, traversal=None):
    if visited is None:
        visited = set()
    if traversal is None:
        traversal = []

    visited.add(start)
    traversal.append(start)

    for child in graph[start]:
        if child not in visited:
            dfs(graph, child, visited, traversal)

    return traversal

# Tree-shaped adjacency list
graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F', 'G'],
    'D': [],
    'E': [],
    'F': [],
    'G': []
}

# Run DFS
order = dfs(graph, 'A')
print("DFS Traversal:", order)

# Create graph in NetworkX
G = nx.DiGraph()
for parent, children in graph.items():
    for child in children:
        G.add_edge(parent, child)

# Fixed positions for exact tree shape
pos = {
    'A': (0, 3),
    'B': (-1, 2),
    'C': (1, 2),
    'D': (-1.5, 1),
    'E': (-0.5, 1),
    'F': (0.5, 1),
    'G': (1.5, 1)
}

# Draw graph
plt.figure(figsize=(6, 4))
nx.draw(G, pos, with_labels=True, node_size=2000, node_color="lightcoral", arrows=True, font_size=12)
plt.title("Tree Structure (DFS)")
plt.show()
```
